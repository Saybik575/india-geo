```python
import pandas as pd
import psycopg2
from psycopg2.extras import execute_values
import os
from pathlib import Path
from dotenv import load_dotenv
```

```python
load_dotenv("backend/.env")
DATABASE_URL = os.getenv("DATABASE_URL")

conn = psycopg2.connect(DATABASE_URL)

cursor = conn.cursor()
```

```python
cursor.execute(
    "INSERT INTO countries (name) VALUES (%s) ON CONFLICT DO NOTHING",
    ("India",),
)
conn.commit()
```

```python
df_states = pd.read_csv('processed_datasets/states.csv')

data = list(df_states.itertuples(index=False, name=None))

query = """
INSERT INTO states (state_code, state_name, country_id)
VALUES %s
ON CONFLICT (state_code) DO NOTHING
"""

execute_values(cursor, query, [
    (row[0], row[1], 1)
    for row in data
])

conn.commit()
```

```python
df_districts = pd.read_csv('processed_datasets/districts.csv')

data = list(df_districts.itertuples(index=False, name=None))

query = """
INSERT INTO districts (district_code, district_name, state_code)
VALUES %s
ON CONFLICT (district_code) DO NOTHING
"""

execute_values(cursor, query, data)

conn.commit()
```

```python
df_subdistricts = pd.read_csv('processed_datasets/subdistricts.csv')

data = list(df_subdistricts.itertuples(index=False, name=None))

query = """
INSERT INTO subdistricts (subdistrict_code, subdistrict_name, district_code)
VALUES %s
ON CONFLICT (subdistrict_code) DO NOTHING
"""

execute_values(cursor, query, data)

conn.commit()
```

```python
df_villages = pd.read_csv('processed_datasets/villages.csv')

data = list(df_villages.itertuples(index=False, name=None))

query = """
INSERT INTO villages (village_code, village_name, subdistrict_code, district_code)
VALUES %s
ON CONFLICT (village_code) DO NOTHING
"""

execute_values(cursor, query, data)

conn.commit()
```

```python
cursor.close()
conn.close()
```
